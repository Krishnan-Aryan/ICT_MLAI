```python
X=df.drop(['price_range'],axis=1)
y=df['price_range']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=3)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
X_train_scale=pd.DataFrame(scale.fit_transform(X_train),columns=X_train.columns,index=X_train.index)
```

```python
X_train_scale
```

```python
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from keras.utils import to_categorical
```

```python
y_train_one_hot = to_categorical(y_train, num_classes=4)
```

#### Hyper Parameter Tuning

```python
def create_model(optimizer='SGD'):
  # create model
  model = Sequential()
  model.add(Dense(12, input_shape=(20,), activation='relu'))
  model.add(Dense(12,  activation='relu'))
  model.add(Dense(4, activation='softmax'))
  # Compile model
  model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
  return model
```

```python
seed=7
tf.random.set_seed(seed)
```

```python
!pip install scikeras
from scikeras.wrappers import KerasClassifier
```

```python
model = KerasClassifier(model = create_model, batch_size = 10 , verbose = 1)
```

```python
# define the grid search parameters
optimizer = ['SGD',  'Adam', 'Adamax', 'Nadam']
epochs = [10,20,30]
learn_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
param_grid = dict(model__optimizer=optimizer,optimizer__learning_rate=learn_rate,epochs=epochs)
```

```python
grid=GridSearchCV(estimator=model,param_grid=param_grid,n_jobs=-1,cv=3)
```

```python
grid_result=grid.fit(X_train_scale,y_train_one_hot)
```

```python
print('Best %f using %s'%(grid_result.best_score_,grid_result.best_params_))
```

```python
X_test_scale=pd.DataFrame(scale.fit_transform(X_test),columns=X_test.columns,index=X_test.index)
y_test_one_hot = to_categorical(y_test, num_classes=4)
```

#### Fitting the Model

```python
from keras.optimizers import Nadam

# Create model
model = Sequential()
model.add(Dense(12, input_shape=(20,), activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(4, activation='softmax'))

# Define optimizer separately
optimizer = Nadam(learning_rate=0.001)

# Compile model
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Train the model
model.fit(X_train_scale, y_train_one_hot, epochs=100, batch_size=32,verbose=1)
```

#### Evaluating the Model

```python
loss, accuracy = model.evaluate(X_test_scale, y_test_one_hot)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```
