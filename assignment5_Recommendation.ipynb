#### Pivot table

```python
user_rating=df.pivot_table(index='user_id',columns='joke_id',values='Rating')
```

```python
user_rating
```

```python
user_rating=user_rating.dropna(thresh=100,axis=1).fillna(0)
```

```python
user_rating.shape
```

```python
user_rating
```

#### correlation method

```python
joke_simil_metrics=user_rating.corr(method='pearson')
joke_simil_metrics.head(40)
```

```python
# making a function to find the similar jocks
```

```python
def get_similar_joks(joke,rating):
    score=joke_simil_metrics[joke]*rating
    score=score.sort_values(ascending=False)
    return score
```

```python
print(get_similar_joks(35,9.438))
```

```python
# recommendation checking 
```

```python
jok_lover=[(1,0.219),(35,9.438),(10,0.229103)]
similar_joks=pd.DataFrame()
for jok,rating in jok_lover:
    similar=similar_joks.append(get_similar_joks(jok,rating))
similar.sum().sort_values(ascending=False)
```

#### By using Neighbours 

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
joke_metrix=csr_matrix(user_rating.values)
```

```python
model=NearestNeighbors(metric='cosine',algorithm='brute')
model.fit(joke_metrix)
```

```python
distances, values = model.kneighbors(joke_simil_metrics.iloc[98,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
print('joke {0} can recommended to users {1}'.format(98,values))
```
